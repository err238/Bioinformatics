#### Bài tập

Sử dụng vòng lặp while () để liên tục nhập và trả kết quả ra màn hình đến khi người dùng nhập N hoặc No hoặc NO thì dừng lại!

ví dụ:  
B1. người dùng nhập chuỗi DNA vào -> trả ra kết quả.  
B2. Hỏi người dùng có muốn tiếp tục không? -> nếu Không thì thoát. Nếu có thì tiếp tục lặp lại bước 1.

```python
def dem_so_nu(dna):
    dna = dna.lower()
    return(dna.count('a'),dna.count('c'),dna.count('g'),dna.count('t'))
```

```python
m = dem_so_nu(dna)
print(m)
```

```python
while(True):
    # cho người dùng nhập chuỗi DNA vào
    dna = input('Mời bạn nhập vào chuỗi DNA: ')
    so_luong_nu = dem_so_nu(dna)
    print(so_luong_nu)
    
    # Tiếp tục chương trình hay không?
    tiep_tuc = input('Bạn có muốn tiếp tục không?').lower()
    if(tiep_tuc == 'n' or tiep_tuc == 'no' or tiep_tuc == 'k' or tiep_tuc == 'khong'):
        print('Cảm ơn bạn đã sử dụng chương trình!')
        break
        
```
