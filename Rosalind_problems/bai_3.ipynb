```python
# chuỗi forward 5' - 3'
f = 'AAAACCCGGT'
```

```python
# A - T T - A
# G - C C - G
# tạo chuỗi reverse
r = '' # chuỗi rỗng

# Kiểm tra và tạo complimentary
for nu in f:
    if(nu == 'A'):
        r = r + 'T'
    elif(nu == 'C'):
        r = r + 'G'
    elif(nu == 'G'):
        r = r + 'C'
    elif(nu == 'T'):
        r = r + 'A'
    else:
        print('Có ký tự không phải là nucleotide!')
        
print(r)
```

```python
r_sau = reversed(r)
r_sau
```

```python
# làm cho chuỗi chạy ngược lại
r_sau = ''.join(reversed(r))
print(r_sau)
```
