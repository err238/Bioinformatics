```python
## Các loại toán tử
### Cộng +
print(3 + 5)
### Trừ -
print(3 - 5)
### Nhân *
print(3 * 5)
### Chia /
print(3 / 5)
### Lũy thừa
print(3 ** 5)
### Chia lấy phần nguyên //
print(7 // 3) # 7/3 = 2 dư 1 -> chỉ lấy 2
### Chia lấy phần dư %
print(7 % 3) # 7/3 = 2 dư 1 -> chỉ lấy 1
```

```python
## Thứ tự tính toán (nên để trong ngoặc đơn)
phep_tinh_1 = 2 + 3 * 5 - 10 / 2
phep_tinh_2 = 2 + (3 * 5) - (10 / 2)
print(phep_tinh_1,phep_tinh_2)
print(type(phep_tinh_1),type(phep_tinh_2))
```

```python
def phep_cong(so_dau_tien, so_thu_hai):
    # thực hiện công so_dau_tien và so_thu_hai
    # trả về giá trị tổng
    tong = so_dau_tien + so_thu_hai
    return tong
```

```python
tong_1 = phep_cong(5, 10)
tong_2 = tong_1 + 10
print(tong_1, tong_2)
```

```python
## các hàm toán học có sẵn
### Giá trị tuyệt đối
so_am = -10
print(abs(so_am))
### Lũy thừa
pow(so_am,2)
### tìm max, min
min(1,3,-1,5)
### làm tròn số
round(10.32554643, 2)
```

```python

## Các hàm với chuỗi
chuoi = " hôm nay tôi hơi mệt! "
print(chuoi)
### xuất ra chiều dài chuỗi
len(chuoi)
### chuyển sang chữ in hoa
chuoi_1 = chuoi.upper() # method phương thức
print(chuoi_1)
### chuyển sang chữ thường
chuoi_2 = chuoi_1.lower()
print(chuoi_2)
### loại bỏ khoảng trắng đầu cuối
chuoi.strip()
### kiểm tra tồn tại
#'nay' in chuoi
#'luoi' in chuoi
'Tôi' in chuoi
### Thay thế 
chuoi.replace(" ","-")
```

```python
## Biến boolean
print(5 == 5) # hai dấu bằng == là bằng, 1 dấu = là gán
### toán tử and or not
print(5 == 5 and 10 < 5)
print(5 == 5 or 10 < 5)
print(not(5 == 5))
```
