```python
bien_toan_cuc = 5 #biến toàn cục hay là biến global
def ham_nao_do():
    bien_trong_ham = 10 #biến cục bộ hay biến local
    print(bien_trong_ham)
    global bien_toan_cuc
    bien_toan_cuc = bien_trong_ham + 20
    print(bien_toan_cuc)

ham_nao_do()
print(bien_toan_cuc)

```
